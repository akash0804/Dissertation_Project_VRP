```python
from scipy.stats import zscore

# Z-score method
z_scores = zscore(df['Total Distance (km) (ResequenceSolution)'])
z_outliers = df[(z_scores > 2.5) | (z_scores < -2.5)]

# IQR method
Q1 = df['Total Distance (km) (ResequenceSolution)'].quantile(0.25)
Q3 = df['Total Distance (km) (ResequenceSolution)'].quantile(0.75)
IQR = Q3 - Q1
iqr_outliers = df[(df['Total Distance (km) (ResequenceSolution)'] < (Q1 - 1.5 * IQR)) | 
                  (df['Total Distance (km) (ResequenceSolution)'] > (Q3 + 1.5 * IQR))]

z_outlier_count = z_outliers.shape[0]
iqr_outlier_count = iqr_outliers.shape[0]

z_outlier_count, iqr_outlier_count

```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

numeric_df = filtered_df.select_dtypes(include=['float64', 'int64'])

# Standardizing the data
standardized_df = (numeric_df - numeric_df.mean()) / numeric_df.std()

nan_values = standardized_df.isnull().sum()
infinite_values = (np.isinf(standardized_df)).sum()

nan_values, infinite_values

# Filling NaN values with 0 for standardized dataset
standardized_df_filled = standardized_df.fillna(0)

# Applying PCA on the filled dataset
pca_filled = PCA(n_components=2)
principal_components_filled = pca_filled.fit_transform(standardized_df_filled)
principal_df_filled = pd.DataFrame(data=principal_components_filled, columns=['PC1', 'PC2'])

# Plotting the 2D projection
plt.figure(figsize=(10, 8))
sns.scatterplot(x=principal_df_filled["PC1"], y=principal_df_filled["PC2"])
plt.title('2D PCA of Adjusted Routes Data (After Handling Missing Values)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np




wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(principal_df_filled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow method
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply KMeans clustering with the optimal number of clusters (based on elbow method result; k=3)
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
clustered_data = kmeans.fit_predict(principal_df_filled)

# Adding cluster information to the principal_df_filled
principal_df_filled['Cluster'] = clustered_data

# Visualize the clusters
plt.figure(figsize=(10,8))
sns.scatterplot(x=principal_df_filled['PC1'], y=principal_df_filled['PC2'], hue=principal_df_filled['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.title('Clusters in PCA Space')
plt.show()

principal_df_filled.to_csv("C:\\Users\\mreth\\Desktop\\Dissertation\\RouteVisualizations3\\principal_df_filled.csv", index=False)
```

```python
from itertools import combinations, groupby
from collections import Counter

# Convert dataset to list of lists for Apriori
records = []
for i in range(len(data)):
    record = [str(data.values[i, j]) for j in range(len(data.columns)) if 'bin' in data.columns[j]]
    records.append(record)

# Define a function to return items and item pairs
def get_item_pairs(order_item):
    order_item = sorted(order_item)
    for item_pair in combinations(order_item, 2):
        yield item_pair

# Define a function to count items and item pairs
def count_items(records):
    item_stats = {}
    pair_stats = {}

    for record in records:
        unique_items = list(set(record))
        pairs = get_item_pairs(unique_items)
        
        for item in unique_items:
            if item in item_stats:
                item_stats[item] += 1
            else:
                item_stats[item] = 1

        for pair in pairs:
            if pair in pair_stats:
                pair_stats[pair] += 1
            else:
                pair_stats[pair] = 1

    return item_stats, pair_stats

item_stats, pair_stats = count_items(records)

# Convert item and pair stats to dataframes
df_item_stats = pd.DataFrame(list(item_stats.items()), columns=['Item', 'Count'])
df_pair_stats = pd.DataFrame(list(pair_stats.items()), columns=['Pair', 'Count'])

df_item_stats = df_item_stats.sort_values(by='Count', ascending=False)
df_pair_stats = df_pair_stats.sort_values(by='Count', ascending=False)

df_pair_stats.head(10)  # Displaying top 10 frequent pairs
```
